```python
import re
import matplotlib.pyplot as plt

# 로그 파일 경로
log_file_path = 'train_005930_a3c_lstm.log'

# 데이터 저장을 위한 리스트
epochs = []
pvs = []

# 로그 파일 읽기
with open(log_file_path, 'r') as file:
    for line in file:
        # 정규 표현식을 사용하여 필요한 데이터 추출
        match = re.search(r'\[Epoch (\d+)/\d+\] .* PV:([\d,]+)', line)
        if match:
            epoch = int(match.group(1))
            pv = int(match.group(2).replace(',', ''))  # 문자열에서 쉼표 제거하고 정수로 변환
            
            epochs.append(epoch)
            pvs.append(pv)

# 그래프 그리기
plt.figure(figsize=(10, 5))
plt.plot(epochs, pvs, label='PV', color='green')
plt.title('PV over Epochs')
plt.xlabel('Epochs')
plt.ylabel('PV')
plt.grid()
plt.legend()

# 그래프 출력
plt.tight_layout()
plt.show()
```
